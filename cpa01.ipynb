This dataset demostrates Human Rights score in different coutries during from 1946-2019. The scores capture the extent to which citizens' physical integrity is protected from government killings, torture, political
imprisonments, extrajudicial executions, mass killings and disappearances. Higher scores mean fewer such abuses. The human rights scores represent the relative position of one country in one specific year relative to the average of all countries across the entire period (1949-2014), which is set to zero. Since the dateset is so big and there are so many coutries involved. I will just pick some countries that I am familar with and interested, say, US, China, etc. 

source: https://ourworldindata.org/human-rights

Interesting questions: 
1. Why, in certain period, China's scores dropped to a valley value?
2. Why, after 1970, there is a surge of US's scores?

# Discussion

1. China
As we can see from the graph of China's human right scores, from 1946 to 1950, the scores continue to drop. During that time, China was under civil war, I persume that would be the reason. And around 1959, there is a sharp valley, to my understanding of China's history, 1959 was the climax of the 'Anti-Rightist Campaign' where the government labelled anyone who potentially undermines its power 'Rightist', which leads to wrongfully convicted cases and human right crisis. During 1965- 1976, there was also a huge valley, that should be due to the 'Cultural Revolution', similar nature to the campaign mentioned above. Lots of people, mostly innocent, were persecuted by the zealous Mao's supporters. Many historians think this whole revolution was about consolidation of power, by suppressing the intellectuals or anyone not 'pro-party' enough, who are more likely to object to the mad dictatorship.  After Mao's death, China's scores got a surge from 1978-1986, however, there is another sharp valley in 1989. From my perspective, that could be due to 'Tian'anmen Square Incident' where liberalists, mostly students, protested in Tian'anmen Square for democracy and political freedom. While they expected revision of constitution and political reform, the government sent troops and tanks as response.

2. US
Though I am not an expert of American history, my best guess is that the surge of scores between 1970-1982 could be due to the victory of Civil Rights Movement and the influence of liberalism, globally popular in the 80s. This surge take US above 0 (the average), all the way above 2. At certain year, the score is higher than China's for 3. What surprises me is, until around 1972, US's scores are below 0, not much higher than China's. I noticed that in 2018 the scores declines to 0 again. 

Some thoughts about the data: Sometimes the score has its relativity. US is a country of complete system of rule of law, and have lots of people focusing on human rights issues, community power, NGOs, etc. But in 2019 the score is only about 1 above China's. Perhaps this is because, when the data was collected, more people in US subjectively feel US have bad Human Rights scores. On the other hand, since China usually label Human Rights topics as sensitive, collecting the data there could be difficult.
